# Blackbody flux

Flux can be calculated with the method ```flux```. It accepts just two arguments - an array of log10(wavelengths) and an array of parameter values (for required parameters - check the ```label_names``` property.)

Wavelengths are expected to be in angstroms, effective temperature in Kelvins, and the output flux is in $\rm{erg}/s/cm^3$.

````python
wavelengths = np.linspace(1000, 40000, 10000)
spectrum_5777 = bb.flux(np.log10(wavelengths), 5777)
````

Let's plot spectra for a few temperature values.

````python
spectrum_4000 = bb.flux(np.log10(wavelengths), 4000)
spectrum_5000 = bb.flux(np.log10(wavelengths), 5000)
spectrum_6000 = bb.flux(np.log10(wavelengths), 6000)

plt.plot(wavelengths, spectrum_4000[:,0], color='firebrick', label='4000 K')
plt.plot(wavelengths, spectrum_5000[:,0], color='yellowgreen', label='5000 K', linestyle='--')
plt.plot(wavelengths, spectrum_5777[:,0], color='royalblue', label='5777 K')
plt.plot(wavelengths, spectrum_6000[:,0], color='deeppink', label='6000 K', linestyle='--')

plt.gca().set_xlabel(r'wavelength [$\AA$]')
plt.gca().set_ylabel(r'flux [$\rm{erg}/s/cm^2$]')
plt.legend()
````
