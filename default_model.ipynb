```python
from tpayne import download_hf_model, TPayne
import matplotlib.pyplot as plt
import numpy as np
```

```python
model = download_hf_model()
```

```python
m = TPayne(model)
```

```python
log_wave = np.linspace(np.log10(4500), np.log10(4600), 1000)
spectral_parameters = np.ones(94)*0.5
mu = 1.0

%time y = m(log_wave, mu, spectral_parameters)
```

```python
%time y = m(log_wave, mu, spectral_parameters)
```

```python
fig, axs = plt.subplots(2, sharex=True)
axs[0].plot(log_wave, y[:,0])
axs[0].plot(log_wave, y[:,1])
axs[1].plot(log_wave, y[:,0]/y[:,1])

axs[0].set_xlabel("log(wavelength)")
axs[0].set_ylabel("Intensity")
axs[1].set_xlabel("log(wavelength)")
axs[1].set_ylabel("Normalized Intensity")
```
